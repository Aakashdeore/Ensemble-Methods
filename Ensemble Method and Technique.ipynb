# Ensemble Methodes/ Techniques

* It is used for larg amount of data set 

* Entire dataset devided into chunks of data set and these chunks run by a seperate model 

* Seperate accuracy of every model combined together and create "Mean Accuracy"

It hase 2 Types 

1) Bagging (parallel Method) -: * RandomForestclassifer(Mostly use)/ 
                                * Extra trees (use for huge dataset)
        
2) Boosting (Serial Method) -: *AdaBoostclassifer/ 
                               *GradientBoostclassifer,/ *votingclassifier

```python
import pandas as pd
import numpy as np 
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
ds=load_iris()
ds
```

```python
df=pd.DataFrame(data=ds.data)
```

```python
df
```

```python
df=pd.DataFrame(data=ds.data,columns=ds.feature_names)
```

```python
df
```

```python
df['target']=pd.DataFrame(ds.target)
df
```

```python
x=df.iloc[:,0:-1]
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

# Bagging 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=100,random_state=42)
#here estimator 100 is default but we can apply estimator 50,100,150,200,250
# the estimator is depend on the hugeness of data. it should no more than 250 
rf.fit(x_train,y_train)
rf_pred=rf.predict(x_test)
print(accuracy_score(y_test,rf_pred))
print(classification_report(y_test,rf_pred))
print(confusion_matrix(y_test,rf_pred))
```

# Boost

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
# N_estimator is default 50 here and we can give it as per the data set like 100,150,200,250
ada=AdaBoostClassifier(n_estimators=50,base_estimator=DecisionTreeClassifier(),learning_rate=1.0)
#  the bracket section is default here
ada.fit(x_train,y_train)
ada_pred=ada.predict(x_test)
print(accuracy_score(y_test,ada_pred))
print(classification_report(y_test,ada_pred))
print(confusion_matrix(y_test,ada_pred))
```

 AdaBoost  in SVC classifier

there are 2 main algorithm in AdaBoost Classifier 

1) Default algorithm is "SAMME.R" -: This algorithm work when SVC has default Kernel ('rbf')
    
2) Algorithm "SAMME" -: This algorithm we have to use when SVC has another kernel like ("Linear","poly")

```python
#when it is default then 
from sklearn.svm import SVC
sv=SVC()
ad=AdaBoostClassifier(n_estimators=100,base_estimator=SVC,algorithm='SAMME.R')
#here default SVC(kernel='rbf') that is why default Adaboost Algorithm is ('SAMME.R')
ad.fit(x_train,y_train)
```

```python

ad_pred=ad.predict(x_test)
print(accuracy_score(y_test,ad_pred))
print(classification_report(y_test,ad_pred))
print(confusion_matrix(y_test,ad_pred))
```
